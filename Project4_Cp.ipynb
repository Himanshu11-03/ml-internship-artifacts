```python
data = pd.merge(df, ds, on='movieId')

data.head()
```

```python
user_movie_matrix = data.pivot_table(
    index='userId',
    columns='title',
    values='rating'
).fillna(0)
user_movie_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(user_movie_matrix)
similarity_df = pd.DataFrame(
    similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)
similarity_df.head()
```

```python
def recommend_movies(user_id, n=5):
    similar_users = similarity_df[user_id].sort_values(
        ascending=False
    )[1:6]
    movies_watched = user_movie_matrix.loc[user_id]
    recommendations = {}
    for sim_user in similar_users.index:
        sim_movies = user_movie_matrix.loc[sim_user]
        for movie, rating in sim_movies.items():
            if movies_watched[movie] == 0 and rating > 4:
                recommendations[movie] = recommendations.get(
                    movie,0
                ) + rating
    recommended = sorted(
        recommendations.items(),
        key=lambda x:x[1],
        reverse=True
    )
    return recommended[:n]
```

```python
recommend_movies(1)
```
