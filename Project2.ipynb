```python
df.fillna(0, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["Inactive"] = df["LastLoginDays"] > 30
```

```python
df["HighComplaints"] = df["Complaints"] > 2
```

```python
df["RiskScore"] = (
    df["LastLoginDays"] * 2 +
    df["Complaints"] * 10
)
```

```python
def risk_category(score):
    if score >= 100:
        return "High"
    elif score >= 50:
        return "Medium"
    else:
        return "Low"
df["RiskCategory"] = df["RiskScore"].apply(risk_category)
```

```python
avg_revenue = df["MonthlyCharges"].mean()
print("Average Revenue:", avg_revenue)
```

```python
premium_count = len(
    df[df["Plan"] == "Premium"]
)
```

```python
print("Premium Customers:", premium_count)
```

```python
high_risk = len(
    df[df["RiskCategory"] == "High"]
)
```

```python
print("High Risk Customers:", high_risk)
```

```python
print(
    df[
        [
            "CustomerID",
            "RiskScore",
            "RiskCategory"
        ]
    ]
)
```

```python
df.to_csv(
    "processed_customers.csv",
    index=False
)
```

```python
print("Processed file saved")
```
